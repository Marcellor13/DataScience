```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dados = pd.read_csv('cidades_br.csv', sep=',', index_col ="UF")
```

```python
dados.head()
```

```python
#01
x= dados.Longitude.values
y= dados.Latitude.values
plt.figure(figsize=(8,8))
plt.scatter(x, y);
```

```python
kmeans = KMeans(n_clusters=26)
estados = kmeans.fit_predict(dados[['Longitude', 'Latitude']].values)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x, y, c=estados);
plt.scatter(kmeans.cluster_centers_[:,0],
 kmeans.cluster_centers_[:,1], c='red');
```

```python
results = dados[['Nome do Município']].copy()
results['Numero de Clusters'] = estados
results.head()
```

```python
#Cidades da região Sudeste

linhasSUD = dados.loc[["ES", "MG", "SP", "RJ"]]
linhasSUD.head()
```

```python
x = linhasSUD.Latitude.values
y = linhasSUD.Longitude.values
plt.figure(figsize=(7, 7))
plt.scatter(x, y);
```

```python
kmeans = KMeans(n_clusters=5)
gruposSUD = kmeans.fit_predict(linhasSUD[['Latitude', 'Longitude']].values)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x, y, c=gruposSUD);
plt.scatter(kmeans.cluster_centers_[:,0],
kmeans.cluster_centers_[:,1], c='orange');
```

```python
results = linhasSUD[['Nome do Município']].copy()
results['Numero de Clusters'] = gruposSUD
results.head()
```

```python
# Quantas cidades temos em cada cluster gerado?

results['Numero de Clusters'].value_counts()
```

```python
#Apresente um histograma da quantidade de cidades por cluster.

u = np.unique(gruposSUD)

plt.figure(figsize=(10,5))
plt.bar(u, [list(gruposSUD).count(i) for i in u]);
plt.xlabel('Clusters')
plt.ylabel('Cidades')
```
